## Missing

```python
#Đọc dữ liệu
dt_mi = pd.read_csv(r"C:\Users\PC\Documents\Khai phá dữ liệu\Data chương 2\Data_Missing_TimeSeries.csv")
dt_mi.info()
```

```python
dt_mi.head()
```

```python
dt_mi.isnull().sum()
```

```python
#Hiển thị dữ liệu missing của 48805
dt_mi.loc[dt_mi[['48805']].isnull().any(axis=1)]
```

```python
#Tính giá trị trung bình
mean1 = round(dt_mi['48805'].mean(),0)
mean2 = round(dt_mi['48806'].mean(),0)
mean3 = round(dt_mi['48825'].mean(),0)
mean4 = round(dt_mi['48838'].mean(),0)
mean5 = round(dt_mi['48839'].mean(),0)
```

```python
#Thay thế các giá trị missing bằng giá trị trung bình
dt_mi['48805'].fillna(mean1, inplace=True)
dt_mi['48806'].fillna(mean2, inplace=True)
dt_mi['48825'].fillna(mean3, inplace=True)
dt_mi['48838'].fillna(mean4, inplace=True)
dt_mi['48839'].fillna(mean5, inplace=True)
dt_mi.info()
```

```python
#Trực quan hóa dữ liệu
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))
tq = dt_mi['48805'].value_counts()

#Vẽ biểu đồ cột 
plt.plot( tq.index, tq.values,  color='#0099CC', marker ='o')
plt.grid(ls='--')
plt.show()
```
