```python
#Khai báo thư viện
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
#Đọc dữ liệu
data = pd.read_csv(r"C:\Users\PC\Documents\Khai phá dữ liệu\Data_Diamond.csv")
data.info()
```

```python
#Hiển thị dữ liệu
data
```

```python
#Quan sát dữ liệu
data.describe()
```

```python
#Trực quan hóa dữ liệu
plt.figure(figsize=(8, 6)) 
plt.scatter(data['Weight(Carat)'],
            data['Price(USD)'],
            c='red')

plt.xlabel('Weight(Carat)')
plt.ylabel('Price(USD)')
plt.grid(True) 
plt.show()
```

```python
#Gọi thuật toán KMeans:
from sklearn.cluster import KMeans

#Khai báo mô hình, với số cụm = 3 và các tham số khác sử dụng mặc định
model = KMeans(n_clusters=3)

#Liệt kê các tham số của mô hình
model.get_params()
```

```python
#Thực hiện phân cụm
model.fit_predict(data)
#Tạo cột Cluster để biết dữ liệu được phân vào cụm nào
data['Cluster'] = model.labels_
print(data)
```
