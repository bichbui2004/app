## Max - Min Scaling

```python
#Tạo bản sao dl
data1 = data.copy()
data1
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler1 = MinMaxScaler()
```

```python
#Xử lý thuộc tính Age
model_age = scaler1.fit(data1[['Age']])
data1['Age'] = model_age.transform(data1[['Age']])

#Xử lý thuộc tính Salary
model_salary = scaler1.fit(data1[['Salary']])
data1['Salary'] = model_salary.transform(data1[['Salary']])

data1
```
