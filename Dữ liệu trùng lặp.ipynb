## Xử lý dữ liệu trùng lặp

```python
#Cách 1: Xóa các hàng trùng lặp
#Tạo bản sao dl
pa1 = data.copy()
#Xóa dữ liệu trùng lặp
pa1 = pa1.drop_duplicates(keep=False)
```

```python
#Hiển thị dl trước và sau xử lý
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div styles="float: left; padding: 10px;">
    <p styles='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(seft, *args):
        seft.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                for a in self.args)
display('data', 'pa1')
```

```python
#Cách 2: Giữ lại hàng đầu tiên
#Tạo bản sao dl
pa2 = data.copy()
#Thực hiện giữ lại bản ghi đầu
pa2 = pa2.drop_duplicates(keep='first')
```

```python
#Hiển thị dữ liệu trước và sau xử lý
display('data','pa2')
```

```python
#Cách 3: Giữ lại hàng cuối cùng, xóa tất cả hàng phía trước
#Tạo bản sao dl
pa3 = data.copy()
#Thực hiện giữ lại bản ghi cuối cùng
pa3 = pa3.drop_duplicates(keep='last')
```

```python
#Hiển thị kết quả trước và sau xử lý:
display('data','pa3')
```

```python
#Cách 4: Đánh dấu các hàng bị trùng lặp
#Tạo bản sao dl
pa4 = data.copy()
#Thực hiện thêm cột Check đánh dấu hàng lặp lại: 1 - lặp lại , 0 - không lặp lại
pa4['Check'] = pa4.duplicated(keep=False).astype(int)
```

```python
#Hiển thị kết quả trước và sau xử lý:
display('data','pa4')
```
