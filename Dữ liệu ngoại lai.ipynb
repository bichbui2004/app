```python
#Đọc dữ liệu
import pandas as pd
data = pd.read_csv(r"C:\Users\PC\Documents\Khai phá dữ liệu\Data chương 2\Data_Outlier.csv")

#Hiển thị tổng quan dữ liệu
data.info()
```

```python
#Sử dụng biểu đồ boxplot để phát hiện dữ liệu ngoại lai:
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
data.boxplot()
plt.ylabel('Nhiệt độ')
plt.show()
```

```python
#Kiểm tra dữ liệu Nha Trang
plt.figure(figsize=(10,4))
plt.plot(data['time'],data['Nha Trang'], 'r--o')
plt.ylabel('Nhiệt độ')
plt.xticks(rotation=90)
plt.grid(ls='--')
plt.show()
```

```python
#--> Các điểm dữ liệu nằm ngoài râu Mininum được xem xét
# là các điểm ngoại lai trái (left outlier)
```

```python
#Kiểm tra dữ liệu Hồ Chí Minh
plt.figure(figsize=(10,4))
plt.plot(data['time'],data['Ho Chi Minh'], 'r--o')
plt.ylabel('Nhiệt độ')
plt.xticks(rotation=90)
plt.grid(ls='--')
plt.show()
```

```python
#--> Các điểm dữ liệu nằm ngoài râu Maxinum được xem xét
# là các điểm ngoại lai phải (right outlier)
```

## Xử lý dữ liệu ngoại lai

```python
#Tạo bản sao dữ liệu
data_outlier = data.copy()
```

```python
#Hiển thị dữ liệu
data_outlier.loc[:10,['time','Nha Trang']]
```

```python
#Với giá trị ngoại lai trạm Nha Trang
#Xưr lý bằng cách thay thế: 2.51 --> 25.1
data_outlier.loc[6,'Nha Trang'] =25.1
```

```python
#Biểu đồ sau khi xử lý dl ngoại lai
plt.figure(figsize=(10,4))

#Dữ liệu cũ:
plt.plot(data['time'], data['Nha Trang'], 'r--o', label = 'Dữ liệu cũ')

#Dữ liệu mới sau khi được xử lý
plt.plot(data_outlier['time'], data_outlier['Nha Trang'], 'b--o', label='Dữ liệu mới')

plt.title('XỬ LÝ DỮ LIỆU NGOẠI LAI TRẠM NHA TRANG')
plt.xlabel('Thời gian (h)')
plt.xticks(rotation=90)
plt.ylabel('Nhiệt độ (oC)')
plt.grid(True)
plt.legend()
plt.show
```

```python
#Hiển thị dữ liệu
data_outlier.loc[:,['time','Ho Chi Minh']]
```

```python
#Với giá trị ngoại lai trạm Hồ Chí Minh
#Xưr lý bằng cách thay thế: 58 --> 28
data_outlier.loc[13,'Ho Chi Minh'] =28
```

```python
#Biểu đồ sau khi xử lý dl ngoại lai
plt.figure(figsize=(10,4))

#Dữ liệu cũ:
plt.plot(data['time'], data['Ho Chi Minh'], 'r--o', label = 'Dữ liệu cũ')

#Dữ liệu mới sau khi được xử lý
plt.plot(data_outlier['time'], data_outlier['Ho Chi Minh'], 'b--o', label='Dữ liệu mới')

plt.title('XỬ LÝ DỮ LIỆU NGOẠI LAI TRẠM HỒ CHÍ MINH')
plt.xlabel('Thời gian (h)')
plt.xticks(rotation=90)
plt.ylabel('Nhiệt độ (oC)')
plt.grid(True)
plt.legend()
plt.show
```
