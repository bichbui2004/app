```python
#Đọc dữ liệu sheet 2010
data_state_population = pd.read_excel(r"C:\Users\PC\Documents\Khai phá dữ liệu\Data chương 2\state_population.xlsx", sheet_name='2010')
print(data_state_population)
```

```python
#Lọc dữ liệu total
data_state_population = data_state_population[data_state_population['ages']=='total']
print(data_state_population)
```

```python
#Lấy cột state/region và population
data_state_population = data_state_population[['state/region','population']]
data_state_population.rename(columns={'state/region':'state'},inplace=True)
data_state_population.reset_index(drop=True,inplace=True)
data_state_population
```

```python
data_state_population.sort_values(by="state")
data_state_population
```

```python
#Đọc file state_areas
data_area = pd.read_csv(r"C:\Users\PC\Documents\Khai phá dữ liệu\Data chương 2\state_areas.csv")
data_area = data_area.sort_values(by="state")
data_area
```

```python
#Đọc file state_code
data_code = pd.read_csv(r"C:\Users\PC\Documents\Khai phá dữ liệu\Data chương 2\state_code.txt",
                       header = None,
                       names = ['Name','Code'])
data_code = data_code.sort_values(by="Code")
data_code
```

```python
#Gộp thành 1 file
data_2010 = pd.concat([data_state_population, data_area, data_code], axis=1)
data_2010
```

```python
data_2010 = data_2010[['Code','Name','population','area (sq. mi)']]
data_2010.reset_index(drop=True,inplace=True)
data_2010
```

```python
data_2010['density'] = np.round(data_2010['population']/data_2010['area (sq. mi)'],1)
data_2010
```

```python
data_2010=data_2010.sort_values(by="density")
```
